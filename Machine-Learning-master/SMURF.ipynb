```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

```python
ageLayer = keras.Input(shape=(1,), name="ageLayer")
bmiLayer = keras.Input(shape=(1,), name="bmiLayer")

denseA = keras.layers.Dense(3, activation="relu", name="denseA")(ageLayer)
denseB = keras.layers.Dense(3, activation="relu", name="denseB")(bmiLayer)

combined = keras.layers.Add(name="combined")([denseA, denseB])
denseCombined = keras.layers.Dense(3, activation="relu", name="denseCombined")(combined)

outputLayer = keras.layers.Dense(1, name="outputLayer")(denseCombined)

#input_layer = keras.Input((2,))
#dense_layer = keras.layers.Dense(4, activation="relu")(input_layer)
#output_layer = keras.layers.Dense(1)(dense_layer)

model = keras.Model([ageLayer, bmiLayer], outputLayer, name="SMURF")
model.summary()
```

```python
keras.utils.plot_model(model, show_shapes=True)
```

```python
model.compile(optimizer="adam", loss="mean_squared_error")
```
